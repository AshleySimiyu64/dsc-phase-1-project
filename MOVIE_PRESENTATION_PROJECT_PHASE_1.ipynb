Real-World Problem
In the entertainment industry, decision-makers such as producers, investors, and marketing teams need to determine which types of films are likely to perform well at the box office. However, without structured data analysis, it’s difficult to recognize long-term trends, compare film performance across years, or identify what content consistently brings in high revenue.
Manually analyzing years of movie revenue data can be time-consuming, inconsistent, and error-prone. As a result, many potentially valuable insights about movie genres, franchises, and audience behavior go unnoticed or are based on intuition rather than data.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data (replace with your real dataset)
top_movies = pd.DataFrame({
    'Movie': [
        'Star Wars: The Force Awakens', 'Avengers: Endgame', 'Black Panther',
        'Avengers: Infinity War', 'Jurassic World', 'The Lion King',
        'The Avengers', 'Frozen', 'Beauty and the Beast', 'Incredibles 2'
    ],
    'Total Domestic Gross (Millions)': [936, 858, 700, 679, 652, 543, 623, 477, 504, 609]
})

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='Total Domestic Gross (Millions)', y='Movie', data=top_movies, palette='viridis')
plt.title('Top 10 Movies by Domestic Gross (2010–2018)')
plt.xlabel('Total Domestic Gross (in Millions USD)')
plt.ylabel('Movie Title')
plt.tight_layout()
plt.show()
```

```python
avg_genre = pd.DataFrame({
    'Genre': ['Action', 'Adventure', 'Animation', 'Comedy', 'Drama', 'Fantasy', 'Sci-Fi'],
    'Average Domestic Gross (Millions)': [580, 520, 400, 280, 210, 450, 470]
})

# Plotting
plt.figure(figsize=(10, 5))
sns.barplot(x='Genre', y='Average Domestic Gross (Millions)', data=avg_genre, palette='magma')
plt.title('Average Domestic Gross by Genre (2010–2018)')
plt.ylabel('Average Domestic Gross (in Millions USD)')
plt.xlabel('Genre')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define problems and their perceived impact scores (1-10 scale)
problems = [
    "High Financial Risk",
    "Uncertain Audience Preferences",
    "Lack of Market Experience",
    "Strong Industry Competition",
    "Poor Timing of Release",
    "No Established Fan Base"
]

impact_score = [9, 8, 8, 7, 6, 6]  # Arbitrary scores to represent relative severity

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(problems, impact_score, color='skyblue')
plt.xlabel("Impact Level (1–10)", fontsize=12)
plt.title("Top Problems Faced by a New Movie Studio", fontsize=14)
plt.gca().invert_yaxis()  # Highest impact at top
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add value labels to bars
for bar in bars:
    plt.text(bar.get_width() + 0.2, bar.get_y() + bar.get_height()/2,
             str(bar.get_width()), va='center', fontsize=10)

plt.tight_layout()
plt.show()
```
