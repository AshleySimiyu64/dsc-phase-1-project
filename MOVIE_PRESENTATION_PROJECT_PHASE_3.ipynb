To help stakeholders in the film industry make informed decisions about which types of films are likely to succeed financially, this dashboard includes three clear, relevant, and well-labeled visualizations. These were designed to be easy to interpret, visually clean, and directly aligned with business value.

1. Bar Chart – Top 10 Grossing Movies (2010–2018)
Purpose: Helps stakeholders quickly see which movies generated the highest domestic revenue over the analysis period.

Stakeholder Value: Highlights high-performing franchises and titles for future investment and marketing insights.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example data (replace with your real dataset)
df_top10 = pd.DataFrame({
    'movie_title': [
        'Star Wars: The Force Awakens', 'Avengers: Endgame', 'Black Panther',
        'Avengers: Infinity War', 'Jurassic World', 'The Lion King',
        'The Avengers', 'Incredibles 2', 'Beauty and the Beast', 'Frozen'
    ],
    'domestic_gross': [936, 858, 700, 679, 652, 543, 623, 609, 504, 477]  # in millions
})

# Sort for clean visual order
df_top10 = df_top10.sort_values(by='domestic_gross', ascending=True)

# Plotting
plt.figure(figsize=(12, 6))
barplot = sns.barplot(
    x='domestic_gross',
    y='movie_title',
    data=df_top10,
    palette='Blues_r'
)

# Title and labels
plt.title('Top 10 Grossing Movies (2010–2018)', fontsize=16)
plt.xlabel('Domestic Gross Revenue (in Millions USD)', fontsize=12)
plt.ylabel('Movie Title', fontsize=12)

# Add value labels to bars
for i, value in enumerate(df_top10['domestic_gross']):
    plt.text(value + 5, i, f"${value}M", va='center', fontsize=10)

# Grid and layout
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data (replace with actual full dataset if needed)
df_top10 = pd.DataFrame({
    'movie_title': [
        'Star Wars: The Force Awakens', 'Avengers: Endgame', 'Black Panther',
        'Avengers: Infinity War', 'Jurassic World', 'The Lion King',
        'The Avengers', 'Incredibles 2', 'Beauty and the Beast', 'Frozen'
    ],
    'release_year': [2015, 2019, 2018, 2018, 2015, 2019, 2012, 2018, 2017, 2013],
    'domestic_gross': [936, 858, 700, 679, 652, 543, 623, 609, 504, 477],
    'international_gross': [1131, 1939, 646, 1369, 1019, 1132, 895, 634, 759, 873]
})

# Group total gross by year (domestic only for this plot)
yearly_gross = df_top10.groupby('release_year')['domestic_gross'].sum().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_gross, x='release_year', y='domestic_gross', marker='o', color='navy', linewidth=2)

# Annotations for standout movies
highlight_movies = df_top10[df_top10['domestic_gross'] > 850]
for _, row in highlight_movies.iterrows():
    plt.text(row['release_year'], row['domestic_gross'] + 15,
             row['movie_title'], fontsize=9, ha='center', color='darkred', wrap=True)

# Trend line (polynomial regression)
z = np.polyfit(yearly_gross['release_year'], yearly_gross['domestic_gross'], 2)
p = np.poly1d(z)
plt.plot(yearly_gross['release_year'], p(yearly_gross['release_year']), linestyle='--', color='gray', label='Trend')

# Labels, titles, polish
plt.title('Total Domestic Gross by Year (Top 10 Movies, 2010–2018)', fontsize=16)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Domestic Gross (in Millions USD)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(np.arange(2010, 2020, 1))
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_top_grossing_movies(df, title_col='movie_title', value_col='domestic_gross'):
    """
    Plot a horizontal bar chart of the top grossing movies.

    Parameters:
    df (DataFrame): DataFrame with movie titles and gross earnings.
    title_col (str): Column name for movie titles.
    value_col (str): Column name for gross earnings.
    """
    # Sort and take top 10
    df_sorted = df.sort_values(by=value_col, ascending=False).head(10)

    # Set figure and plot
    plt.figure(figsize=(12, 6))
    sns.barplot(x=value_col, y=title_col, data=df_sorted,
                palette='Blues_r', edgecolor='black')

    # Add labels and title
    plt.title('Top 10 Grossing Movies (2010–2018)', fontsize=14)
    plt.xlabel('Domestic Gross (in Millions USD)')
    plt.ylabel('Movie Title')
    plt.grid(axis='x', linestyle='--', alpha=0.6)

    # Add value labels
    for index, value in enumerate(df_sorted[value_col]):
        plt.text(value + 5, index, f"${value:.0f}M", va='center', fontsize=9)

    plt.tight_layout()
    plt.show()

# Example usage
plot_top_grossing_movies(df_top10)
```
